```python
import pandas as pd
tienda1 = pd.read_csv("challenge1-data-science-latam/base-de-datos-challenge1-latam/tienda_1 .csv")
```

```python
tienda2 = pd.read_csv("challenge1-data-science-latam/base-de-datos-challenge1-latam/tienda_2.csv")
```

```python
tienda3 = pd.read_csv("challenge1-data-science-latam/base-de-datos-challenge1-latam/tienda_3.csv")
```

```python
tienda4 = pd.read_csv("challenge1-data-science-latam/base-de-datos-challenge1-latam/tienda_4.csv")
```

```python
# 1. Análisis de facturación
```

```python
facturacion1 = tienda1["Precio"].sum()
facturacion2 = tienda2["Precio"].sum()
facturacion3 = tienda3["Precio"].sum()
facturacion4 = tienda4["Precio"].sum()

print("Facturación Tienda 1:", facturacion1)
print("Facturación Tienda 2:", facturacion2)
print("Facturación Tienda 3:", facturacion3)
print("Facturación Tienda 4:", facturacion4)
```

```python
# 2. Ventas por categoría
```

```python
tienda1.columns
```

```python
print("Tienda 1")
print(tienda1["Categoría del Producto"].value_counts())

print("Tienda 2")
print(tienda2["Categoría del Producto"].value_counts())

print("Tienda 3")
print(tienda3["Categoría del Producto"].value_counts())

print("Tienda 4")
print(tienda4["Categoría del Producto"].value_counts())
```

```python
# 3. Calificación promedio de la tienda
```

```python
calificacion1 = tienda1["Calificación"].mean()
calificacion2 = tienda2["Calificación"].mean()
calificacion3 = tienda3["Calificación"].mean()
calificacion4 = tienda4["Calificación"].mean()
```

```python
print("Calificación promedio Tienda 1:", calificacion1)
print("Calificación promedio Tienda 2:", calificacion2)
print("Calificación promedio Tienda 3:", calificacion3)
print("Calificación promedio Tienda 4:", calificacion4)
```

```python
print("Calificación promedio Tienda 1:", round(calificacion1,2))
print("Calificación promedio Tienda 2:", round(calificacion2,2))
print("Calificación promedio Tienda 3:", round(calificacion3,2))
print("Calificación promedio Tienda 4:", round(calificacion4,2))
```

```python
# 4. Productos más y menos vendidos
```

```python
# Contar cuántas veces se vendió cada producto
ventas_tienda1 = tienda1["Producto"].value_counts()
ventas_tienda2 = tienda2["Producto"].value_counts()
ventas_tienda3 = tienda3["Producto"].value_counts()
ventas_tienda4 = tienda4["Producto"].value_counts()
```

```python
print("Tienda 1 - Más vendido:", ventas_tienda1.idxmax(), "con", ventas_tienda1.max(), "ventas")
print("Tienda 1 - Menos vendido:", ventas_tienda1.idxmin(), "con", ventas_tienda1.min(), "ventas")

print("Tienda 2 - Más vendido:", ventas_tienda2.idxmax(), "con", ventas_tienda2.max(), "ventas")
print("Tienda 2 - Menos vendido:", ventas_tienda2.idxmin(), "con", ventas_tienda2.min(), "ventas")

print("Tienda 3 - Más vendido:", ventas_tienda3.idxmax(), "con", ventas_tienda3.max(), "ventas")
print("Tienda 3 - Menos vendido:", ventas_tienda3.idxmin(), "con", ventas_tienda3.min(), "ventas")

print("Tienda 4 - Más vendido:", ventas_tienda4.idxmax(), "con", ventas_tienda4.max(), "ventas")
print("Tienda 4 - Menos vendido:", ventas_tienda4.idxmin(), "con", ventas_tienda4.min(), "ventas")
```

```python
# 5. Envío promedio por tienda
```

```python
envio1 = tienda1["Costo de envío"].mean()
envio2 = tienda2["Costo de envío"].mean()
envio3 = tienda3["Costo de envío"].mean()
envio4 = tienda4["Costo de envío"].mean()
```

```python
print("Costo de envío promedio Tienda 1:", round(envio1,2))
print("Costo de envío promedio Tienda 2:", round(envio2,2))
print("Costo de envío promedio Tienda 3:", round(envio3,2))
print("Costo de envío promedio Tienda 4:", round(envio4,2))
```
